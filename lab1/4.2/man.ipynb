```python
class Patient:
    def __init__(self, medical_record_number):
        self.medical_record_number = medical_record_number
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, medical_record_number):
        new_patient = Patient(medical_record_number)
        if self.is_empty():
            self.front = self.rear = new_patient
        else:
            self.rear.next = new_patient  # 新病人排在队尾
            self.rear = new_patient  # 更新队尾
        print(f"病人 {medical_record_number} 已排队。")

    def dequeue(self):
        if self.is_empty():
            print("没有病人排队。")
            return None
        patient = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        print(f"病人 {patient.medical_record_number} 正在就诊。")
        return patient.medical_record_number

    def print_queue(self):
        if self.is_empty():
            print("没有病人排队。")
            return
        current = self.front
        print("当前排队的病人病历号：")
        while current:
            print(current.medical_record_number, end=" ")
            current = current.next
        print()

def hospital_sys():
    queue = Queue()
    while True:
        print("\n请选择操作：")
        print("1. 排队")
        print("2. 就诊")
        print("3. 查看排队情况")
        print("4. 下班")
        choice = input("输入选择（1/2/3/4）：")

        if choice == '1':
            medical_record_number = input("请输入病人的病历号：")
            queue.enqueue(medical_record_number)
        elif choice == '2':
            queue.dequeue()
        elif choice == '3':
            queue.print_queue()
        elif choice == '4':
            print("下班了，再见！")
            break
        else:
            print("无效选择，请重新输入。")

def main():
    hospital_sys()

if __name__ == "__main__":
    main()
```
